```python
import numpy as np
from sklearn.model_selection import train_test_split, RandomizedSearchCV
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import pandas as pd

# Split data into Training (75%) and Testing (25%)
x_t, x_te, y_t, y_te = train_test_split(x, y, test_size=0.25, random_state=20)

# Initialize KNN and find the best K using RandomizedSearchCV
knn = KNeighborsClassifier()
params = {'n_neighbors': list(np.arange(2, 50))}
nknn = RandomizedSearchCV(knn, n_jobs=-1, random_state=20, scoring='accuracy', param_distributions=params, cv=10)
nknn.fit(x_t, y_t)

# Display the best parameters and accuracy score
print("Best K:", nknn.best_params_)
print("Best Training Score:", nknn.best_score_)
nknn = nknn.best_estimator_

# Prediction on training and testing sets
pred_t = nknn.predict(x_t)
pred = nknn.predict(x_te)
print('KNN Training Accuracy:', accuracy_score(y_t, pred_t))
print('KNN Testing Accuracy:', accuracy_score(y_te, pred))
print('\nClassification Report for KNN Classifier:\n\n', classification_report(y_te, pred))

# Confusion Matrices
cmk_t = confusion_matrix(y_t, pred_t)
cmk = confusion_matrix(y_te, pred)

# Prediction function for new data
def predict_lung_cancer(input_data):
    """
    Predicts whether or not a person has lung cancer based on input parameters.

    Parameters:
        input_data (list): Input features for the new sample.

    Returns:
        str: Prediction result ('Has Lung Cancer' or 'No Lung Cancer')
    """
    # Convert input data to a DataFrame to match training format
    input_df = pd.DataFrame([input_data], columns=x.columns)

    # Predict using the best estimator
    prediction = nknn.predict(input_df)

    # Interpret the prediction
    if prediction[0] == 1:
        return "Has Lung Cancer"
    else:
        return "No Lung Cancer"

# Example usage
input_data = [
    33,  # Age
    1,   # Gender
    6,   # Air Pollution
    6,   # Alcohol use
    6,   # Dust Allergy
    6,   # Occupational Hazards
    6,   # Genetic Risk
    6,   # Chronic Lung Disease
    6,   # Balanced Diet
    6,   # Obesity
    3,   # Smoking
    2,   # Passive Smoker
    6,   # Chest Pain
    6,   # Coughing of Blood
    6,   # Fatigue
    3,   # Weight Loss
    4,   # Shortness of Breath
    2,   # Wheezing
    2,   # Swallowing Difficulty
    3,   # Clubbing of Finger Nails
    1,   # Frequent Cold
    2,   # Dry Cough
    3    # Snoring
]
print("Prediction for new data:", predict_lung_cancer(input_data))
```
